## Setup

### Importanto bibliotecas:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Carregando o dataset:

```python
df = pd.read_csv('dataset1.csv')
```

### Informações básicas sobre o dataset:

```python
df.info()
df.describe()
```

## Questão A

```python
#Convertendo para formato datetime:  
df['data'] = pd.to_datetime(df['data'].str.split('-', expand=True)[1].str.strip(), format='%d/%m/%Y %H:%M')

#Criando uma coluna com o dia da semana:
df['dia_semana'] = df['data'].dt.day_name()
df.sort_values(by='data', inplace=True)

print(df)
```

## Questão B

```python
#Plot da temperatura por tempo:
plt.figure(figsize=(15, 5))
sns.scatterplot(x=df["data"], y=df["temperatura"])
plt.title('Temperatura por tempo')
plt.xlabel('Data')
plt.ylabel('Temperatura')
plt.xticks(rotation=45)
plt.show()
```

**Conclusões:**

Analisando o gráfico, podemos observar que há vários valores ausentes nas medições de temperatura ao longo do tempo. Além disso, podemos notar um padrão interessante: a temperatura atinge valores mais elevados durante o período do meio do mês, enquanto no final do mês, a temperatura tende a ser mais baixa em comparação com o início do mês. 

## Questão C

```python
# Criando um dicionário de cores para cada pessoa
cores_pessoas = {'Fábio': 'blue', 'Fernando': 'red', 'Marcela': 'green', 'Rafael': 'orange', 'Júlia': 'purple', 'Letícia': 'brown'}


# Plotando os gráficos:
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))
for i, pessoa in enumerate(pessoas):
    dados_pessoa = df[df['responsável'] == pessoa]
    cor_pessoa = cores_pessoas.get(pessoa)
    
    ax = axes[i // 2, i % 2]
    ax.scatter(dados_pessoa['data'], dados_pessoa['temperatura'], color=cor_pessoa)

    ax.set_title(f'Temperatura por Tempo - {pessoa}')
    ax.set_ylabel('Temperatura (°C)')
    ax.set_xlabel('Data e Hora')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

**Conclusões:**

Após uma análise cuidadosa dos dados fornecidos pelos alunos, foi observado que a característica que poderia levantar mais suspeitas é o fato de que o aluno "Fernando" é o único que possui todas as medições de temperatura para o seu horário designado. Diferentemente dos demais alunos, Fernando não apresenta valores ausentes em suas medições, o que pode levantar suspeitas sobre a integridade dos dados ou o padrão de coleta de informações para esse indivíduo.

Quanto às estatísticas de médias, mínimos, máximos e desvios, todos os alunos parecem possuir dados condizentes e coerentes entre si. Portanto, em termos de estatísticas gerais, não há indícios de comportamentos suspeitos entre os alunos.

## Questão D

Realizando uma interpolação linear com os dados:

```python
#Copiando o dataframe para interpolá-lo:
df_interpol = df.copy()

#Definindo a coluna data como index:
df_interpol.set_index('data', inplace=True)

#Realizando a  interpolação linear:
df_interpol = df_interpol.interpolate(method='time')

#Visualizando o dataframe interpolado:
print(df_interpol)

#Gerando um gráfico com os dados interpolados:
plt.figure(figsize=(15, 5))
sns.lineplot(x=df_interpol.index, y=df_interpol["temperatura"])
plt.title('Temperatura por tempo (dados interpolados)')
plt.xlabel('Data')
plt.ylabel('Temperatura')
plt.xticks(rotation=45)
plt.show()
```

## Questão F

```python
#Extraindo o dia da data e calculando as temperaturas máximas e mínimas de cada dia
df['dia'] = df['data'].dt.day
temperaturas_maximas = df.groupby('dia')['temperatura'].max()
temperaturas_minimas = df.groupby('dia')['temperatura'].min()

#Criando o gráfico com as temperaturas máximas e mínimas:
plt.figure(figsize=(10, 6))

# Plotando o traçado vermelho (temperaturas máximas)
plt.plot(temperaturas_maximas.index, temperaturas_maximas.values, 'r-', label='Temperatura Máxima')

# Adicionando as marcas para a maior temperatura do traçado vermelho
dia_maxima = temperaturas_maximas.idxmax()
temperatura_maxima = temperaturas_maximas.max()
plt.text(dia_maxima, temperatura_maxima, str(temperatura_maxima), ha='right', va='bottom', color='red')

# Plotando o traçado azul (temperaturas mínimas)
plt.plot(temperaturas_minimas.index, temperaturas_minimas.values, 'b-', label='Temperatura Mínima')

# Adicionando as marcas para a menor temperatura do traçado em azul:
dia_minima = temperaturas_minimas.idxmin()
temperatura_minima = temperaturas_minimas.min()
plt.text(dia_minima, temperatura_minima, str(temperatura_minima), ha='right', va='top', color='blue')

# Adicionando linhas pontilhadas para a maior e menor temperatura:
plt.axhline(temperatura_maxima, color='red', linestyle='--')
plt.axhline(temperatura_minima, color='blue', linestyle='--')

plt.xlabel('Dia')
plt.ylabel('Temperatura (°C)')
plt.title('Temperaturas Máximas e Mínimas ao Longo do Mês')
plt.legend()
plt.grid(True)
plt.xticks(range(1, 32))
plt.show()
```
